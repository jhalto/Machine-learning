```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

loading data

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/snails_gc.csv')
```

Visulization

```python
df.head(15)
```

Handle Categorical data

```python
df['M'] = df['M'].astype('category')
df['M'] = df['M'].cat.codes
```

```python
df.head()
```

checking if there is any null value

```python
df.isnull().sum()
```

preparing data for training

```python
X= df.drop(columns='M',axis=1)
X
```

```python
y = df['M']
y
```

spilliting_data

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.10,random_state=0)
```

```python
lr =LogisticRegression()
```

apply Logistic regression model

```python
lr.fit(X_train,y_train)
```

Evulating

```python
y_pred = lr.predict(X_test)
con_m= metrics.confusion_matrix(y_test,y_pred)
accuracy = metrics.accuracy_score(y_test,y_pred)
classi = metrics.classification_report(y_test,y_pred)

print(con_m)
print(classi)
print(accuracy)
```
